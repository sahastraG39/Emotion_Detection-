This project detects the emotion present in a given text
using a pre-trained AI model.

```python
from transformers import pipeline
```

```python
emotion_classifier = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    return_all_scores=True
)
```

```python
text = "I can't believe you did this to me."
```

```python
results = emotion_classifier(text)
results
```

```python
top_emotion = max(results[0], key=lambda x: x['score'])
top_emotion
```

```python
texts = [
    "I am so happy today!",
    "This is the worst day of my life.",
    "I am afraid of what will happen next.",
    "Wow! I never expected this."
]

for t in texts:
    result = emotion_classifier(t)
    top = max(result[0], key=lambda x: x['score'])
    print(f"Text: {t}")
    print(f"Detected Emotion: {top['label']} ({top['score']:.2f})\n")
```
